```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Criação do DataFrame
data = {
    'Dia': ['Segunda-Feira', 'Terça-Feira', 'Quarta-Feira', 'Quinta-Feira', 'Sexta-Feira', 'Sábado', 'Domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

df = pd.DataFrame(data)

#DataFrame
print(df)

#Dados do Relatório

#Total de Horas Trabalhadas
total_horas = df['Horas Trabalhadas'].sum()
print(f"Total de Horas Trabalhadas: {total_horas}")

#Média Diária de Horas Trabalhadas
media_horas = df['Horas Trabalhadas'].mean()
print(f"Média Diária de Horas Trabalhadas: {media_horas:.2f}")

#Total de Bugs Corrigidos
total_bugs = df['Bugs Corrigidos'].sum()
print(f"Total de Bugs Corrigidos: {total_bugs}")

#Média Diária de Bugs Corrigidos
media_bugs = df['Bugs Corrigidos'].mean()
print(f"Média Diária de Bugs Corrigidos: {media_bugs:.2f}")

#Total de Tarefas Concluídas
total_tarefas = df['Tarefas Concluídas'].sum()
print(f"Total de Tarefas Concluídas: {total_tarefas}")

#Média Diária de Tarefas Concluídas
media_tarefas = df['Tarefas Concluídas'].mean()
print(f"Média Diária de Tarefas Concluídas: {media_tarefas:.2f}")

df['Produtividade Diária'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas'] #Produtividade Diária (Tarefas Concluídas por Hora)


# Visualizações
plt.figure(figsize=(14, 10))

# Gráfico de barras agrupadas para Horas Trabalhadas, Bugs Corrigidos e Tarefas Concluídas
ax = df.set_index('Dia')[['Horas Trabalhadas', 'Bugs Corrigidos', 'Tarefas Concluídas']].plot(kind='bar', figsize=(14, 8), color=['blue', 'orange', 'green'])
plt.title('Horas Trabalhadas, Bugs Corrigidos e Tarefas Concluídas por Dia')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()

# Gráfico de linha para Produtividade Diária
sns.lineplot(x='Dia', y='Produtividade Diária', data=df, marker='o', color='blue')
plt.title('Produtividade Diária (Tarefas Concluídas por Hora)')
plt.xticks(rotation=45)
```
